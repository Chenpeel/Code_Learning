```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi(n):
    r = 1
    count_in_circle = 0
    x_in_circle = []
    y_in_circle = []
    for i in range(n):
        x = random.uniform(0, r)
        y = random.uniform(0, r)
        if math.sqrt(x**2 + y**2) <= r:
            count_in_circle += 1
            x_in_circle.append(x)
            y_in_circle.append(y)
        else:
            continue
    
    x_out_circle = r * np.cos(np.arange(0, np.pi/2, 0.01))
    y_out_circle = r * np.sin(np.arange(0, np.pi/2, 0.01))
    fig, ax = plt.subplots()
    ax.plot(x_out_circle, y_out_circle)
    ax.scatter(x_in_circle, y_in_circle, color='red', s=1)
    ax.set_aspect('equal', adjustable='box')
    plt.show()

    pi = 4 * count_in_circle / n
    return pi

n1 = 1000000
n2 = 10000000
# n3 = 1000000000
# 数据过大导致内存占满，终止

pi1 = estimate_pi(n1)
pi2 = estimate_pi(n2)
# pi3 = estimate_pi(n3)

print("投掷总数为10E5时，pi值为:", pi1)
print("投掷总数为10E7时，pi值为:", pi2)
# print("投掷总数为10E9时，pi值为:", pi3)
```
